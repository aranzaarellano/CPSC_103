```python
##################
# Data Definitions for gdp_data.csv

Gdp = str # in range [0, ...]
#  interpret the GDP (Gross Domestic Product) of a country

GDP1 = "13000897.71"
GDP2 = "34568999.0"
GDP3 = "0"

@typecheck
#template based on atomic non-distinct
def fn_for_gdp(g: Gdp) -> ...:
    return ...(g)



# List[Gdp]
# interp. a list of GDP data

LOG0 = [GDP1]
LOG1 = [GDP1, GDP2]
LOG2 = []

@typecheck
# template based on arbitrary-sized data and reference rule
def fn_for_log(log: List[Gdp]) -> ...:
    # description of the acculumator
    acc = ... #type: ...
    
    for g in log:
        acc = ...(acc, fn_for_gdp(g))
        
    return acc

Country = Enum("Country", ["Can", "Aus", "Dnk", "Fra", "Mex", "Ind", "Bra", "Phl"])
# interp. a country, it can either be Canada(Can), Australia (Aus), Denmark (DNK),
# France (Fra), Mexico(Mex), India (Ind), Brazil (Bra), or Philippines (PHL)

# examples are redundant for enumerations

@typecheck
#template from one-of (8 cases), and 8 cases are atomic non-distinct
def fn_for_country(c: Country) -> ...:
    if c == Country.Can:
        return ...
    elif c == Country.Aus:
        return ...
    elif c == Country.DNK:
        return ...
    elif c == Country.Fra:
        return ...
    elif c == Country.Mex:
        return ...
    elif c == Country.Ind:
        return ...
    elif c == Country.Bra:
        return ...
    elif c == Country.Phl:
        return ...


    
    
Gdp_Info = NamedTuple("Gdp_Info", [("country", Country), ("gdp", List[Gdp])])
# interp. the Gdp Info, which includes the name of the country and a list of it's GDP for 
# the years 1960 to 2017

GDP_1 = Gdp_Info("Can", LOG0)
GDP_2 = Gdp_Info("Aus", LOG1)

#template from Compound
def fn_for_gdp_info(g: Gdp_Info) -> ...:
    return ...(fn_for_country(g.country),
              fn_for_log(g.gdp))


#List[Gdp_Info]
# interpret a list of GDP information

LOGI1 = [GDP_1, GDP_2]
LOGI2 = [GDP_1]
LOGI3 = []

@typecheck
#template from arbitrary-sized and reference rule
def fn_for_logi(logi: List[Gdp_Info]) -> ...:
    #description of acc
    acc = ... #type: ...
    
    for gi in logi:
        acc = ...(acc, fn_for_gdp_info(gi))
    return acc


Gdp_Growth = float
#interpret the GDP growth rate

GDPG1 = 1.2
GDPG2 = -2.30
GDPG3 = 0

@typecheck
# template based on Atomic Non-distinct
def fn_for_gdp_growth(gg: Gdp_Growth) -> ...:
    return ...(gg)

#List[Gdp_Growth]
# interpret a list of Gdp growth rates

L1 = [GDPG1, GDPG2, GDPG3]
L2 = [GDPG3]
L3 = []

@typecheck
# template based on arbitrary sized and reference rule
def fn_for_lgg(lgg: List[Gdp_Growth]) -> ...:
    #description of acc
    acc = ... #type: ...
    
    for gg in lgg:
        acc = ...(acc, fn_for_gdp_growth(gg))
    return acc
    


Data = str
# interp. if there is data or no data (none)
                                     
D1 = "1956"
D2 = "1956", "258777"
D3 = ".."

@typecheck
#template based on atomic non-distinct
def fn_for_data(d: Data) -> ...:
    if d is none:
        return ...
    else: 
        return ...(d)

    
    
# List[Data]
# interp. a list of data

LOD1= [D1, D2]
LOD2= [D1, D3]
LOD3= []

@typecheck
#template based on arbitrary-sized and refrence rule
def fn_for_lod(lod: List[Data]) -> ...:
    #description of accumulator
    acc = ... #type: ...
    
    for d in lod:
        acc = ...(d, acc)
    return acc
```

```python
@typecheck
def main(filename_exp_data: str, filename_gdp_data: str, c1: Country, c2: Country) -> None:
    """
    Reads the file from given filename, analyzes the data,
    returns a graph plotting the compounded gdp growth rates and compounded expenditure growth rates for 
    two given countries, against each other
    """
    # return None # stub
    # Template from HtDAP, based on composition
    
    return plot_growth_rates(compound_growth(expenditure_growth(read2(filename_exp_data), c1)), 
                             compound_growth(gdp_growth(read1(filename_gdp_data), c1)),
                             compound_growth(expenditure_growth(read2(filename_exp_data), c2)),
                             compound_growth(gdp_growth(read1(filename_gdp_data), c2)), c1, c2)
```

```python
## Analyze and it's helper functions
@typecheck
def plot_growth_rates(expenditure_growth: List[Exp_Growth], gdp_growth: List[Gdp_Growth], expenditure_growth1: List[Exp_Growth], gdp_growth2: List[Gdp_Growth], c1: Country, c2: Country) -> None:
    """
    plot the Gdp growth rate and government expenditure of two countries against each other
    """
    #return None #stub
    #template from visualization and Composition
    
    pyplot.xlabel("Year")
    pyplot.ylabel("Growth Rate")
    pyplot.title(plot_name_2(c1, c2))
    
    expenditure_plot = pyplot.plot(years(expenditure_growth), expenditure_growth, label = label_name_expenditure(c1))
    gdp_plot = pyplot.plot(years(gdp_growth), gdp_growth, label = label_name_gdp(c1))
    
    expenditure_plot1 = pyplot.plot(years(expenditure_growth1), expenditure_growth1, label = label_name_expenditure(c2))
    gdp_plot1 = pyplot.plot(years(gdp_growth2), gdp_growth2, label = label_name_gdp(c2))
    
    pyplot.setp(gdp_plot, color = "blue", marker = "o")
    pyplot.setp(expenditure_plot, color= "orange", marker="o" )
    
    pyplot.setp(gdp_plot1, color = "green", marker = "o")
    pyplot.setp(expenditure_plot1, color= "yellow", marker="o" )
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    
    pyplot.grid(True)
    
    pyplot.show()
    
    return None
    
@typecheck
def label_name_expenditure(c: Country) -> str:
    """
    return the name of the expenditure label of data for the graph, of the given Country
    """
    #return "" #stub
    #template from Country
    
    return plot_name(c)[20: ]
    
@typecheck
def label_name_gdp(c: Country) -> str:
    """
    return the name of the GDP label of data for the graph, of the given Country
    """
    #return "" #stub
    #template from Country
    
    return plot_name(c)[0: 15] + plot_name(c)[43: ]

    
@typecheck
def years(lof: List[float]) -> List[float]:
    """
    returns a list of years for a list in float
    """
    #return [] #stub
    # template based on arbitrary-sized
    
    #acc stores the years in loe seen so far
    acc = []
    
    beg_year = 1975
    
    for e in range(0, len(lof)):
        year = beg_year + e
        acc.append(year)
    return acc
        

@typecheck
def plot_name(c1: Country) -> str:
    """
    returns the name of the plot, that has the name of the given Country 1 included
    """
    #return "" #stub
    #template from Country
    
    if c1 == Country.Can:
        return "GDP growth rate vs. Expenditure growth rate for Canada"
    elif c1 == Country.Aus:
        return "GDP growth rate vs. Expenditure growth rate for Australia"
    elif c1 == Country.Dnk:
        return "GDP growth rate vs. Expenditure growth rate for Denmark"
    elif c1 == Country.Fra:
        return "GDP growth rate vs. Expenditure growth rate for France"
    elif c1 == Country.Mex:
        return "GDP growth rate vs. Expenditure growth rate for Mexico"
    elif c1 == Country.Ind:
        return "GDP growth rate vs. Expenditure growth rate for India"
    elif c1 == Country.Bra:
        return "GDP growth rate vs. Expenditure growth rate for Brazil"
    elif c1 == Country.Phl:
        return "GDP growth rate vs. Expenditure growth rate for Philippines"

@typecheck
def plot_name_2(c1: Country, c2: Country) -> str:
    """
    return the name of the plot that has the name of the given Country 1 and 2 included
    """
    #return "" #stub
    #template from Country
    
    if c2 == Country.Can:
        return plot_name(c1) + " And Canada"
    elif c2 == Country.Aus:
        return plot_name(c1) + " And Australia"
    elif c2 == Country.Dnk:
        return plot_name(c1) + " And Denmark"
    elif c2 == Country.Fra:
        return plot_name(c1) + " And France"
    elif c2 == Country.Mex:
        return plot_name(c1) + " And Mexico"
    elif c2 == Country.Ind:
        return plot_name(c1) + " And India"
    elif c2 == Country.Bra:
        return plot_name(c1) + " And Brazil"
    elif c2 == Country.Phl:
        return plot_name(c1) + " And Philippines"
    

@typecheck
def expenditure_growth(loe: List[Expenditure_Info], c: Country) -> List[Exp_Growth]: 
    """
    return a list of annual expenditure growth rates
    """
    #return [] #stub
    # template from List[Expenditure_Info] with additional parameter
    
    #growth rate keeps track of the expenditure growth rates in loe seen so far
    growth_rate = []
    
    for Expenditure_Info in loe:

        if Expenditure_Info.country == is_country(c):
            
                for a in range(1, len(Expenditure_Info.exp)):
                    grates = ((Expenditure_Info.exp[a] - (Expenditure_Info.exp[a - 1])) / (Expenditure_Info.exp[a - 1])) * 100 
                    growth_rate.append(grates)
    
    return growth_rate


@typecheck
def compound_growth(lof: List[float]) -> List[float]:
    """
    returns a list of compounded growth rates
    """
    # return []
    # template from List[float]
    
    #acc keeps track of the compunded growth rates in lof seen so far
    acc = []
    number = 0
    
    for f in lof:
        x = number + f
        number = x
        acc.append(x)
    return acc 


@typecheck
def gdp_growth(log: List[Gdp_Info], c: Country) -> List[Gdp_Growth]: 
    """
    return a list of annual Gdp growth rates for a given country
    """
    # return [] #stub
    # template from List[Gdp_Info] with additional parameter
    
    #growth_rate keeps track of the annual Gdp growth rates seen so far
    growth_rate = []
    
    for Gdp_Info in log:

        if Gdp_Info.country == is_country(c):
            
                for a in range(1, len(Gdp_Info.gdp)):
                    grates = ((Gdp_Info.gdp[a] - (Gdp_Info.gdp[a - 1])) / (Gdp_Info.gdp[a - 1])) * 100 
                    growth_rate.append(grates)
    
    return growth_rate

@typecheck
def is_country(c: Country) -> str:
    """
    return a tring representing the World Bank code for one of the
    given countries
    """
    # return "" #stub
    # template from Country
    
    if c == Country.Can:
        return "CAN"
    elif c == Country.Aus:
        return "AUS"
    elif c == Country.Dnk:
        return "DNK"
    elif c == Country.Fra:
        return "FRA"
    elif c == Country.Mex:
        return "MEX"
    elif c == Country.Ind:
        return "IND"
    elif c == Country.Bra:
        return "BRA"
    elif c == Country.Phl:
        return "PHL"
```

```python
###########
# Read Functions for gdp_data.csv

@typecheck
def read1(filename: str) -> List[Gdp_Info]:
    """    
    reads information from the specified file and returns a list of GDP data
    """
    # return []  #stub
    # Template from HtDAP
    # loc contains the result so far
    loc = [] # type: List[Consumed]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        
        for row in reader:
            if is_valid(row) == True:
                 c = Gdp_Info(row[3], parse_gdp(row[19: 59]))
             
                 loc.append(c)
    
        return loc

@typecheck
def is_valid(lod: List[Data]) -> bool:
    """
    return True if the data in the list is complete, and return False, if the data in the list is incomplete
    """
    # return True #stub
    # template from List[Data]
    
    for d in lod:
        if d == "..":
            return False
    return True

@typecheck
def parse_gdp(log: List[Gdp]) -> List[float]:
    """
    takes in a list of GDP (in strings), and converts it into a list of floats
    """
    #return [0.0] #stub
    # template from List[GDP]
    
    #acc keeps track of the string-formatted-GDP data that has been converted into float, seen so far
    acc = [] #type: List[float]
    
    for g in log:
        a = float(g)
        acc.append(a)
    return acc

###########
# Read Functions for for expenditure_data.csv

@typecheck
def read2(filename: str) -> List[Expenditure_Info]:
    """    
    reads information from the specified file and returns a list of government expenditure information
    """
    # return []  #stub
    # Template from HtDAP
    # loc contains the country and its expenditure information seen so far
    loc = [] # type: List[Expenditure_Info]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        
        for row in reader:
            if is_valid(row) == True:
                 c = Expenditure_Info(row[3], parse_exp(row[4: 59]))
             
                 loc.append(c)
    
        return loc
    
    
    
@typecheck
def parse_exp(loe: List[Expenditure]) -> List[float]:
    """
    takes in a list of government expenditures (formatted in strings) and converts it into floats
    """
    #return [0.00] #stub
    #template from List[Expenditure]
    
    #acc keeps track of the list of floats
    acc = [] #type: List[float]
    
    for e in loe:
        a = float(e)
        acc.append(a)
    return acc

```
